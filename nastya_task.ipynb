```python
# Сделать бы загрузку сразу боекомплекта для отгрузки - словарь, 
# в котором список модулей по отгрузке и их количество по всей отгрузке.
# например 2 pca-10-1000 и 3 pca-40-800
# каждое значение словаря умножить на число блоков
# объединить два словаря, а значения сложить.

# moduls in q blocks
def moduls_in_block(moduls_dict, q_blocks):
    return {k: v*q_blocks for k,v in moduls_dict.items()}
```

```python
# moduls in order (in all blocks)
def moduls_in_all_block(all_block_list):
    moduls_dict = {} 
    for d in all_block_list:
        for k,v in d.items():
            moduls_dict.update({k:v}) if k not in moduls_dict.keys() else moduls_dict.update({k:v + moduls_dict.get(k)})
    return moduls_dict
```
